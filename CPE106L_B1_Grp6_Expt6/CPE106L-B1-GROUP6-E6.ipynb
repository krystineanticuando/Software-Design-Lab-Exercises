```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def simplePlotSimpler(data_np):
    plt.figure()
    totals = data_np.sum(axis=1)   
    plt.plot(totals, 'bx', label = 'totals') 
    plot=plt.plot(data_np,  '1')
    plt.legend(iter(plot),('problem1','pr2','pr3','pr4'))

def simplePlot(data_np):
    totals = data_np.sum(axis=1)   
    plt.plot(totals, 'bx', label = 'totals')
    clrs = ('green','orange', 'purple', 'red')
    
    for col in range(data_np.shape[1]):        
        plt.plot(data_np[:, col], '1', color = clrs[col], 
                 label = 'problem' + str(col+1))
         
    plt.grid()
    plt.title('Midterm grades')
    plt.legend()
    return 

def sortedPlot (data_np):
    plt.figure()
    dframe = pd.DataFrame(data_np, 
                          columns = ['problem'+str(i) for i in range(1,5)])
    dframe['totals'] = dframe.sum(axis = 1)    
    dframe.sort_values(by='totals', ascending=False, inplace = True)
  
        
    dframe.index = range(1, len(dframe.index)+1)
    print (dframe) 
    
    clrs = ['green','orange', 'purple', 'red', 'blue']
    for col in dframe.columns:
       plt.plot (  dframe[col], '1', color = clrs.pop(), label = col)
 
    plt.yticks(range(0, 101, 5) )
    plt.grid(color = 'lightgrey', linestyle = '--', linewidth = .5)
    plt.ylabel('grade')
    plt.xlabel ('students')
    plt.title('Midterm grades')
    plt.legend()
 

def main():
    data = np.loadtxt('grades3.txt')
    data2 = np.loadtxt('grades4.txt')
    data = np.concatenate( (data,data2) )
    print (data)
    simplePlot(data)
    simplePlotSimpler(data)
    sortedPlot(data)
    plt.show()

main()
```
